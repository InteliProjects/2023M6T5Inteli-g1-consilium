```python
import pandas as pd
```

```python
df_servicos = pd.read_csv("../dados/servicos_setembro_pr_curitiba.txt")
df_tecnicos = pd.read_csv("../dados/tecnicos_setor_pr_curitiba.txt", sep=";")
```

#**Serviços**

```python
df_servicos.shape
```

```python
df_servicos.head()
```

```python
df_servicos_date = df_servicos.copy()
df_servicos_date["dt_fim_execucao"] = pd.to_datetime(df_servicos_date["dt_fim_execucao"])

count_per_day = df_servicos_date.groupby(df_servicos_date['dt_fim_execucao'].dt.date).size()
```

```python
df_timeSeries= pd.DataFrame(count_per_day)
```

```python
!pip install pycaret
```

```python
from pycaret.time_series import *
s = setup(df_timeSeries.reset_index()[0], fh = 1, session_id = 73)
```

```python
check_stats()
```

```python
best = compare_models()
```

```python
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 2})
```

```python
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 15})
```
